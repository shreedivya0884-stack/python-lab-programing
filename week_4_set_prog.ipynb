```python
#set elements to check the program is printing duplicate elemenets
numbers={1,2,3,4,5}
print("Set:",numbers)
```

```python
#To adding elemets
numbers={1,2,3,5,6}
numbers.add(4)
print("Set:",numbers)
```

```python
#to removing element
numbers={1,2,3,4}
numbers.discard(2)
print("Set:",numbers)
```

```python
#to adding more elements
numbers={1,2,5,6}
numbers.add(3)
numbers.add(4)
print("Set:",numbers)
```

```python
#to removing more elements
numbers={1,2,3,4,5,6,7,8}
numbers.discard(2)
numbers.discard(5)
print("Set:",numbers)
```

```python
#set operations
evens={2,4,6}
print("Union:",numbers|evens)
print("Intersection:",numbers&evens)
```
