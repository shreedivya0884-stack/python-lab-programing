```python
#create list and display the list
fruit=["apple","banana","cherry","apple"]
print("Original list:",fruit)
```

```python
#add element
fruit=["apple","banana"]
fruit.append("orange")
print("Original list:",fruit)
```

```python
#remove element
fruit=["apple","banana"]
if "banana" in fruit:
    fruit.remove("banana")
    print("Original list:",fruit)
```

```python
#access by index to print first element
fruit=["apple","banana"]
print("first fruit:",fruit[0])
```

```python
#access by index to print last element
fruit=["apple","banana","orange"]
print("last fruit:",fruit[-1])
```

```python
#list comprehension
fruit=["apple","orange"]
upper_fruit=[fruit.upper() for fruit in fruit]
print("Upper case fruits:",upper_fruit)
```

```python
#for lower case
fruit=["APPLE","ORANGE"]
lower_fruit=[fruit.lower() for fruit in fruit]
print("lower case fruit:",lower_fruit)
```

```python
#find the sum of all elements in [5,10,15,20]
list=[5,10,15,20]
total=0

for num in list:
    total+=num
print("sum:",total)
```

```python
#find maximum and minimum in [8,3,9,1,6]
numbers=[8,3,9,1,6]
maximum=max(numbers)
minimum=min(numbers)
print("Maximum:",maximum)
print("Minimum:",minimum)
```
